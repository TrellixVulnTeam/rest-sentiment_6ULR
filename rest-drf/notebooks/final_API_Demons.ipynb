## Token Authentication

### Token Login

```python
TokenLoginData = {
    "username": "amirberenji",
    "password": "amir1374",
}

TokenLogin = requests.post('http://127.0.0.1:8000/auth/token/login/', TokenLoginData)
token = TokenLogin.json()
print(token)
```

```python
while True:
    TokenLogin = requests.post('http://127.0.0.1:8000/auth/token/login/', TokenLoginData)
    print(TokenLogin, '\n\n', TokenLogin.json())
```

### Token Logout

```python
TokenLogoutData = {
    'Authorization': "Token " + f'{token["auth_token"]}'
}

TokenLogout = requests.post('http://127.0.0.1:8000/auth/token/logout/', headers = TokenLogoutData)
TokenLogout
```
